# Example Structure

Tiap orang boleh memiliki struktur yang berbeda-beda. Software engineering sudah seperti seni bagi para programmer agar code tersusun dengan rapi dan maintainable sehingga masing-masing programmer juga memiliki style masing-masing.

Berikut contoh yang bisa kamu coba untuk projek kamu
```
project
    assets
    data
    output
    notebook
    src
        __init__.py
        model.py
        dataset.py
        utils.py
        exception.py
        predict.py
    config.py
    train.py
    environment.yml
    README.md
    LICENSE
```

Berikut keterangan mengenai struktur tersebut
- `assets`<br>
folder ini berisi file media (biasanya image) yang digunakan pada README dan notebook.
- `data`<br>
folder ini berisi dataset untuk training
- `output`<br>
folder untuk output model, log, dll.
- `notebook`<br>
folder untuk menyimpan code berupa jupyter notebook. Tentunya di dalam jupyter perlu menambahkan PATH apabila ingin digunakan. Hubungi saya untuk info lebih lanjut mengenai ini karena tidak saya bahas disini
- `src`<br>
Folder yang isinya source code. Ada banyak source yang mungkin kamu code, tapi berikut yang umum dipakai di PyTorch:
    - `__init__.py`<br>
    file yang dipakai python untuk menginisialisasi folder menjadi module
    - `model.py`<br>
    berisi arsitektur custom yang digunakan untuk training
    - `dataset.py`<br>
    berisi custom Dataset dan Dataloader
    - `utils.py`<br>
    berisi fungsi-fungsi utilitas
    - `exception.py`<br>
    berisi custom exception yang dibutuhkan pada source code lainnya
    - `predict.py`<br>
    berisi class yang dapat melakukan prediksi data baru
- `config.py`<br>
Tempat untuk config yang boleh diedit oleh pengguna
- `train.py`<br>
berisi pytorch training workflow (dataset & dataloader, config, training preparation MCOC, dan training)
- `environment.yml`<br>
file untuk membuat conda environment projek ini
- `README.md`<br>
Berisi deskripsi projek, instruksi untuk instalasi, training, dan cara penggunaan lainnya.
- `LICENSE`<br>
Lisensi projek ini. Cek lebih lanjut di https://help.github.com/en/github/creating-cloning-and-archiving-repositories/licensing-a-repository

Tentunya seiring berjalannya waktu kamu bisa improvisasi struktur seperti apa yang sesuai untuk projekmu.
